```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine

    
```

```python
engine = create_engine('postgresql://postgres:{password}@localhost:5432/SQL homework')

```

```python

conn = engine.connect()
```

```python
salary_data = pd.read_sql("select * from salaries", conn)

salary_data.head()
```

```python
dm_data = pd.read_sql("select * from dept_manager", conn)

dm_data.head()
```

```python
title_data = pd.read_sql("select * from titles", conn)

title_data.head()
```

```python
employee_data = pd.read_sql("select * from employee", conn)
employee_data.rename(columns={'emp_title_id' : 'title_id'},inplace=True)
employee_data.head()
```

```python
salary_data.head()
```

```python
merged_data = pd.merge(title_data, employee_data, on="title_id", how="inner" )

merged_data.head()
```

```python
merged_data_2 = pd.merge(merged_data, salary_data, on="emp_no", how="inner" )
merged_data_2.head()
```

```python
grouped_data = merged_data_2.groupby("title").mean()
grouped_data
```

```python
revised_data = grouped_data.drop(columns = "emp_no")
revised_data
```

```python
revised_data = revised_data.reset_index()
revised_data
```

```python
x_axis = revised_data["title"]
y_axis = revised_data["salary"]
ticks = np.arange(len(x_axis))
plt.bar(x_axis, y_axis, align = "center", alpha=.5)

plt.xticks(ticks, x_axis, rotation = "vertical")
plt.xlabel=("Title")
plt.ylabel=("Salary")
plt.title=("Average Salary by Title")
plt.show()
```
